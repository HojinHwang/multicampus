```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
tips.tail() 
```

```python
tips
```

```python
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips,
            palette='Set3')
plt.show()
```

```python
g = sns.FacetGrid(tips, col='time', size=5, aspect=.7)
g.map(sns.boxplot, 'sex', 'total_bill', 'smoker').despine(left=True).add_legend(title='smoker')
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', row='sex', col='time', data=tips, size=3)
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights = flights.pivot('month', 'year', 'passengers')
```

```python
flights.head()
```

```python
sns.heatmap(flights)
```

```python
sns.heatmap(flights, annot=True, fmt='d')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.set(style='ticks', color_codes=True)
```

```python
sns.pairplot(iris, hue='species')
```

```python
sns.pairplot(iris, vars=['sepal_width', 'sepal_length'], size=5)
```

```python
sns.pairplot(iris, diag_kind='kde')
```

```python
sns.pairplot(iris, kind='reg', hue='species')
```

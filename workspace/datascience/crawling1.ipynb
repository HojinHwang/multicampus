```python
# pip install BeautifulSoup4
# pip install lxml
# lxml 이 html.parser 보다 성능이 좋다!
from bs4 import BeautifulSoup
from urllib.request import urlopen

url="http://movie.naver.com/movie/sdb/rank/rmovie.nhn"
soup = BeautifulSoup(urlopen(url).read(), 'lxml')
```

```python
# <div class="tit3">
#     <a href="/movie/bi/mi/basic.nhn?code=136315" title="어벤져스: 인피니티 워">어벤져스: 인피니티 워</a>
# </div>
# rank_list = soup.findAll("div", "tit3")

rank_list = soup.select("div.tit3")
rank_list
```

```python
count = 1
for i in rank_list:
    title = i.find('a').string
    print('{0:2d}위: {1}'.format(count, title))
    count = count + 1
```

```python
for idx, rank in enumerate(rank_list, 1):
    title = rank.find('a').string
    print('{0:2d}위: {1}'.format(idx, title))
```
